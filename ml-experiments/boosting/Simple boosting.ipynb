### A simple boosting implementation
 - Simple Decision Tree regressor based on fast.ai implementation!
 - Gradient Boosting using the residual: $r_i = y - y\text{_}pred$, which in practice will minimize the $MSE = \sum_{i=1}^{N}(Y_i - Y\text{_}pred_i)^2$
